```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from PIL import Image
import imageio
import numpy as np
```

```python
model = tf.keras.models.load_model("model_final_improved.h5")
```

```python
filename = "aahad.png"
```

```python
img = Image.open(filename)
img
```

```python
fileResized = filename[:-4] + "Resized.png"
```

```python
img = img.resize((48,48))
img = img.convert("L")
img.save(fileResized)
img
```

```python
img = imageio.imread(fileResized)
img = img.reshape(1,48,48)
img = img/255.0
img
```

```python
prediction_arr = model.predict(img)

print("Probabilities of Each Category: ", prediction_arr)

category = np.argmax(prediction_arr)

dictionary = {0:'Angry', 1:'Disgust', 2:'Fear', 3:'Happy', 4:'Sad', 5:'Surprise', 6:'Neutral'}

result = dictionary[category]
print()
print()
print("This Image represents the Emotion", result)
```
